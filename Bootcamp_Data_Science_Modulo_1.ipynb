## Ordenação e série temporal

Agora que temos a coluna total, vamos ordenar descendente pelo total e depois remover essa coluna total pra não estragar o grafico.

```python
usaveis.head()
ordenados_por_total = usaveis.sort_values("Total", ascending=False)
ordenados_por_total = ordenados_por_total.drop("Total", axis=1, errors="ignore")
ordenados_por_total.head()
```

**Desafio 01: Escolher uma palete de cores mais adequada do matplotlib.**


```python
n_colors = len(ordenados_por_total)
colors = plt.cm.nipy_spectral(np.linspace(0, 1, n_colors))
grafico = ordenados_por_total.T.plot(figsize=(12,8), color=colors)
grafico.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #posiciona a legenda do lado de fora
grafico.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}")) #formata os valores do eixo Y
plt.title("Valor total gasto nos atendimentos por unidade da federação")
plt.show()

```

**Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Sudeste e Centro-Oeste) de cada estado.**

```python
# Map de valores por regiao
regions_map = {
           '1' : 'Norte',
           '2' : 'Nordeste',
           '3' : 'Sudeste',
           '4' : 'Sul',
           '5' : 'Centro-Oeste'
}
regions = [regions_map[idx[0]] for idx in ordenados_por_total.index] #pega o primeiro caracter do index e associa ao map para pegar a região

ordenados_por_total["Região"] = regions
ordenados_por_total.head()
```

**Desafio 03: Formatar o gráfico de custos por mês dos 5 estados, deixando ele agradável (Bonitão, segundo o Gui)**

```python
n_colors = len(ordenados_por_total)
colors = plt.cm.nipy_spectral(np.linspace(0, 1, n_colors))
dados_sem_discrepancia = ordenados_por_total[:5].T[6:-1] #pega menos meses para não mostrar os dados que pioram muito a visualização. Remove o ultimo que é a Região.
grafico = dados_sem_discrepancia.plot(figsize=(12,8), color=colors)
grafico.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #posiciona a legenda do lado de fora
grafico.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}")) #formata os valores do eixo Y
plt.title("Valor total gasto nos atendimentos por unidade da federação")
plt.show()
```

Eles falam bastante sobre as possibilidades de manipulação das informações. Se pegamos apenas alguns ranges de datas ou se não estabelecemos os limites dos eixos, principalmente o eixo y com os valores totais, pode deixar a entender dados errados.

**Desafio 01: Refaça as manipulações feitas em aulas e gaste um tempo analisando os dados e entendendo os efeitos das manipulações**

Exemplo, se não se estabelece o "ymin" para o grafico, ao comparar os dados de Paraná e São Paulo, podemos ter uma ideia que o gasto de SP chega a 4x o do PR.

```python


n_colors = len(ordenados_por_total)
colors = plt.cm.nipy_spectral(np.linspace(0, 1, n_colors))
dados_sem_discrepancia = ordenados_por_total[:3].T[6:-1] #pega menos meses para não mostrar os dados que pioram muito a visualização. Remove o ultimo que é a Região.
dados_sem_discrepancia = dados_sem_discrepancia / 1000000
grafico = dados_sem_discrepancia.plot(figsize=(12,8), color=colors)
grafico.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #posiciona a legenda do lado de fora
grafico.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}")) #formata os valores do eixo Y
plt.ylabel('Valor total (em milhões)', fontsize=15)
plt.title("Valor total gasto nos atendimentos por unidade da federação")
plt.show()
```

Note que nesse gráfico acima, os dados praticamente começam em 50 milhões, o que pode deixar uma visão torta ao apenas olhar para o gráfico. Caso a pessoa leia os dados, consegue interpretar melhor. Mas é sempre melhor para nós cientistas de dados deixar essas coisas claras. Com isso, é melhor usar o "ylim"

```python
n_colors = len(ordenados_por_total)
colors = plt.cm.nipy_spectral(np.linspace(0, 1, n_colors))
dados_sem_discrepancia = ordenados_por_total[:3].T[6:-1] #pega menos meses para não mostrar os dados que pioram muito a visualização. Remove o ultimo que é a Região.
dados_sem_discrepancia = dados_sem_discrepancia / 1000000
grafico = dados_sem_discrepancia.plot(figsize=(12,8), color=colors)
grafico.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #posiciona a legenda do lado de fora
grafico.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}")) #formata os valores do eixo Y
plt.ylabel('Valor total (em milhões)', fontsize=15)
plt.title("Valor total gasto nos atendimentos por unidade da federação")
plt.ylim(0, 400)
plt.show()
```

Note que estabelecendo os limites em Y, o gráfico fica mais correto visualmente, sem permitir possíveis erros de interpretação. 

Outras possibilidades de manipulações é criar um gráfico mais apertado na horizontal e outro mais largo. O que pode dar a impressão no primeiro, que ocorre uma aceleração maior dos valores do que a versão mais larga.

```python
dados_sem_discrepancia = ordenados_por_total[:3].T[6:-1]
dados_sem_discrepancia = dados_sem_discrepancia / 1000000
grafico = dados_sem_discrepancia.plot(figsize=(8,8), color=colors) # Note como a inclinação fica maior do que a do grafico anterior.
```
