Analisando os dados do grafico acima, percebemos que neste mês de Março de 2010 temos um gasto com procedimentos hospitalares em SP mais do que o dobro do segundo estado. Acredito que essa diferença se deva pelo tamanho da população maior em SP.

# Aula 2

```python
import pandas as pd
```

```python
dados = pd.read_csv("/content/A113926189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",")
dados.head()
```

```python
dados[["2008/Ago", "2008/Set"]].head()
```

Neste DataFrame, temos algumas colunas, principalmente de anos anteriores, com dados em branco, e uma forma de excluirmos essas colunas é utilizando o comando mean(), que só traz as medias que tenham alguma dado. Com isso, é possível notar abaixo que ao executar, traz apenas as colunas com dados.

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

O set_index é usado para setar uma coluna como o indice do DataFrame

```python
usaveis = usaveis.set_index("Unidade da Federação")
usaveis.head()
```

```python
usaveis["2019/Ago"].head()
```

O comando abaixo é usado para localizar uma linha especifica na tabela

```python
usaveis.loc["12 Acre"]
```

O comando abaixo é usado para rotacionar a tabela


```python
usaveis.T.head()
```

```python
usaveis.T.plot(figsize=(10,6))
```

Note que o grafico acima não ficou legal. Isso ocorre porque é necessario retirar do DataFrame o total.

```python
usaveis = usaveis.drop("Total", axis=1)
```

```python
usaveis.T.plot(figsize=(10,6))
```

Agora sim, sem os totais, o grafico ficou melhor de se ver pra cada unidade da federação.

**Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada.**

```python
grafico = usaveis.T.plot(figsize=(12,8))
grafico.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #posiciona a legenda do lado de fora
grafico.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}")) #formata os valores do eixo Y
plt.title("Valor total gasto nos atendimentos por unidade da federação")
plt.show()
```

Utilizei o bbox_to_anchor para posicionar do lado de fora do gráfico. Utilizei o link abaixo como referência:
https://matplotlib.org/2.0.2/users/legend_guide.html

**Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência**

```python
estados_selecionados = ["41 Paraná", "31 Minas Gerais", "33 Rio de Janeiro", "53 Distrito Federal", "13 Amazonas"]
grafico = usaveis.T.loc[:, estados_selecionados].plot(figsize=(12,6))
grafico.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #posiciona a legenda do lado de fora
grafico.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}")) #formata os valores do eixo Y
plt.title("Valor total gasto nos atendimentos por unidade da federação")
plt.show()
```

Filtrei os estados escolhidos utilizando a função loc, passando uma variavel com os estados selecionados.
