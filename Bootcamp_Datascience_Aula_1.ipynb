Caderno de exploração dos dados do datasus 2008 pra frente


```python
import pandas as pd
```

```python
dados = pd.read_csv("/content/A123647189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",")
```

```python
dados.head()
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
dados.mean()
```

```python
dados["2008/Ago"].mean()
```

```python
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(15,7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(15,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Internações por unidade da federação do mês de Agosto de 2008")
plt.show()
```
